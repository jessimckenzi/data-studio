```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('pr_1901_2015.xls')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df = df[df['\tYear'] > 1982] 
df.head()
```

```python
df = df.pivot_table(index="\tYear")
df.head()
```

```python
df.dtypes
```

```python
df.plot(y='pr',legend=False)
```

```python
df.to_csv('annual_rainfall.csv', index=False)
```
