```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
# from selenium.webdriver.support.ui import WebDriverWait
```

```python
driver = webdriver.Chrome()
driver.get('http://origin.cpc.ncep.noaa.gov/products/analysis_monitoring/ensostuff/ONI_v5.php')
```

```python
nino = driver.find_element_by_xpath('/html/body/table[4]/tbody/tr/td[2]/table[2]')
print(nino)
```

```python
nino.text
```

```python
rows = nino.find_elements_by_tag_name('tr')[1:-1]
table = []
for row in rows:
    nino_year = {}
    nino_year['year'] = row.find_elements_by_tag_name('td')[0].text
    nino_year['DJF'] = row.find_elements_by_tag_name('td')[1].text
    nino_year['JFM'] = row.find_elements_by_tag_name('td')[2].text
    nino_year['FMA'] = row.find_elements_by_tag_name('td')[3].text
    nino_year['MAM'] = row.find_elements_by_tag_name('td')[4].text
    nino_year['AMJ'] = row.find_elements_by_tag_name('td')[5].text
    nino_year['MJJ'] = row.find_elements_by_tag_name('td')[6].text
    nino_year['JJA'] = row.find_elements_by_tag_name('td')[7].text
    nino_year['JAS'] = row.find_elements_by_tag_name('td')[8].text
    nino_year['ASO'] = row.find_elements_by_tag_name('td')[9].text
    nino_year['SON'] = row.find_elements_by_tag_name('td')[10].text
    nino_year['OND'] = row.find_elements_by_tag_name('td')[11].text
    nino_year['NDJ'] = row.find_elements_by_tag_name('td')[12].text
    print("Scraping next row")
    table.append(nino_year)
```

```python
df = pd.DataFrame(table)
df
```

```python
df = df[df.year != 'Year']
```

```python
df.dtypes
```

```python
df.AMJ = df.AMJ.astype(float)
```

```python
df.ASO = df.ASO.astype(float)
df.DJF = df.DJF.astype(float)
df.FMA = df.FMA.astype(float)
df.JAS = df.JAS.astype(float)
df.JFM = df.JFM.astype(float)
df.JJA = df.JJA.astype(float)
df.MAM = df.MAM.astype(float)
df.MJJ = df.MJJ.astype(float)
df.NDJ = df.NDJ.astype(float)
df.OND = df.OND.astype(float)
df.SON = df.SON.astype(float)
df.year = df.year.astype(int)
```

```python
df.dtypes
```

```python
df.to_csv("nino_years.csv", index=False)
```
