```python
for trail in df:
    df['HIKING'] = df.TRLUSE.str.extract(r"([Hh]ike)")
```

```python
for trail in df:
    df['PackOrSaddle'] = df.TRLUSE.str.extract(r"(Pack.*Saddle)")
```

```python
df.PackOrSaddle.value_counts()
```

```python
df.PackOrSaddle = df.PackOrSaddle.str.replace('Pack and Saddle', 'PackOrSaddle')
```

```python
df.head()
```

```python
for trail in df:
    df['BIKING'] = df.TRLUSE.str.extract(r"(\b[Bb]i\w+e\b)")
```

```python
df.BIKING.value_counts()
```

```python
df.BIKING = df.BIKING.str.replace('Bicycle', 'Bike')
```

```python
for trail in df:
    df['ATV'] = df.TRLUSE.str.extract(r"(\bA.*T.*V)")
```

```python
df.ATV.value_counts()
```

```python
df.ATV = df.ATV.str.replace("All-Terrain Vehicle\|Four-Wheel Drive V","ATV")
```

```python
df.ATV = df.ATV.str.replace("All-Terrain Vehicle\|OHV","ATV")
```

```python
df.ATV = df.ATV.str.replace("All-Terrain V","ATV")
```
