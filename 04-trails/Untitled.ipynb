```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
```

```python
driver = webdriver.Chrome()
driver.get('http://www.peakbagger.com/report/report.aspx?r=w')
```

```python
isolated = driver.find_element_by_xpath('//*[@id="RGridView"]')
print(isolated)
```

```python
isolated.text
```

```python
rows = isolated.find_elements_by_tag_name('tr')[1:]
table = []
for row in rows:
    details = {}
    details['complex'] = row.find_elements_by_tag_name('td')[0].text
    print(details['complex'])
    details['state'] = row.find_elements_by_tag_name('td')[1].text
    print(details['state'])
    details['radius'] = row.find_elements_by_tag_name('td')[2].text
    print(details['radius'])
    details['area'] = row.find_elements_by_tag_name('td')[3].text
    print(details['area'])
    details['center'] = row.find_elements_by_tag_name('td')[4].text
    print(details['center'])
    details['lat'] = row.find_elements_by_tag_name('td')[5].text
    print(details['lat'])
    details['long'] = row.find_elements_by_tag_name('td')[6].text
    print(details['long'])
    print("Scraping next row")
    table.append(details)
```

```python
df = pd.DataFrame(table)
df.head(100)
```

```python
df.to_csv("isolated.csv", index=False)
```

```python
df.area = df.area.astype('float', inplace=True)
df.lat = df.lat.astype('float', inplace=True)
df.long = df.long.astype('float', inplace=True)
df.radius = df.radius.astype('float', inplace=True)
```

```python
df.to_csv("isolated.csv", index=False)
```
